```python
from pathlib import Path

def validate_pipeline_inputs():
    data_dir = Path('data')
    reports_dir = Path('reports')

    required_source_files = [
        data_dir / 'existing_pacemaker_manufacturing.txt',
        data_dir / 'proposed_LVAD_requirements.txt',
    ]
    required_result_files = [
        reports_dir / 'zero_shot_result.txt',
        reports_dir / 'enhanced_zero_shot_result.txt',
        reports_dir / 'cot_result.txt',
    ]

    missing_files = [p for p in (required_source_files + required_result_files) if not p.exists()]
    if missing_files:
        missing_text = '\n'.join(f'- {p.as_posix()}' for p in missing_files)
        raise FileNotFoundError(f'Missing required file(s):\n{missing_text}')

    empty_result_files = [p for p in required_result_files if p.stat().st_size == 0]
    if empty_result_files:
        empty_text = '\n'.join(f'- {p.as_posix()}' for p in empty_result_files)
        raise ValueError(f'Result file(s) are empty:\n{empty_text}')

    return required_source_files, required_result_files

try:
    source_files, result_files = validate_pipeline_inputs()
    print('Safety checks passed. Starting comparison pipeline...')

    source_docs = {
        'existing': source_files[0].read_text(encoding='utf-8'),
        'proposed': source_files[1].read_text(encoding='utf-8'),
    }
    report_docs = {p.stem: p.read_text(encoding='utf-8') for p in result_files}

    print(f"Loaded source documents: {', '.join(source_docs.keys())}")
    print(f"Loaded result reports: {', '.join(report_docs.keys())}")
except Exception as e:
    print(f'Pipeline safety check failed: {e}')
```
