# Topic 1: For Loop

### Generation Prompt
> I have five customers at Cred Pay with IDs from cid_0001 to cid_0005. Can you write a Python program that prints each Customer ID one by one?

```python
# Printing each Customer ID from cid_0001 to cid_0005

for i in range(1, 6):
    customer_id = f"cid_{i:04d}"
    print(customer_id)
```

### Explanation

The code above uses a **for loop** to repeat a block of code five times. Here is what happens step by step:

1. `range(1, 6)` generates a sequence of numbers: 1, 2, 3, 4, 5. The starting number is 1 and it stops just before 6.
2. On each pass through the loop, the variable `i` takes the next number from that sequence.
3. `f"cid_{i:04d}"` is a formatted string. The `:04d` part pads the number with leading zeros so it always has four digits (e.g., 1 becomes 0001).
4. `print()` displays the resulting Customer ID on the screen.

**Key takeaway:** Instead of writing five separate print statements, the for loop handles the repetition automatically.

### Explanation Prompt
> Can you explain this code in simple terms? How does it know to print five IDs and stop?

```python
# Let's trace through the loop to see what happens on each pass

for i in range(1, 6):
    customer_id = f"cid_{i:04d}"
    print(f"Pass {i}: i = {i}, Customer ID = {customer_id}")

print("\nThe loop has finished. It ran exactly 5 times because range(1, 6) produces numbers 1 through 5.")
```

### Explanation

The loop knows to stop because `range(1, 6)` defines exactly how many numbers to produce. Once `i` has taken the value 5 (the last number before 6), the loop completes and Python moves on to whatever comes after it.

Think of `range(1, 6)` as a numbered ticket dispenser: it hands out tickets numbered 1 through 5, then runs out. Each ticket triggers one pass through the loop.

### Modification Prompt
> The company just got 3 more customers, so now we have IDs up to cid_0008. Can you update the code to print all eight?

```python
# Updated: Printing Customer IDs from cid_0001 to cid_0008

for i in range(1, 9):
    customer_id = f"cid_{i:04d}"
    print(customer_id)
```

### Explanation

The only change was updating `range(1, 6)` to `range(1, 9)`. Since `range` stops just before the second number, using 9 means it now produces numbers 1 through 8. This is the power of loops: scaling from 5 customers to 8 (or 8,000) requires changing just one number.

# Topic 2: While Loop

### Generation Prompt
> At Cred Pay, we process customers one at a time. Can you write a program that starts with customer cid_0001 and keeps printing the next customer ID until it finishes cid_0005? But this time, I want it to keep going based on a condition rather than a fixed count.

```python
# Processing customers one at a time using a condition based approach

counter = 1

while counter <= 5:
    customer_id = f"cid_{counter:04d}"
    print(customer_id)
    counter = counter + 1

print("\nAll customers have been processed.")
```

### Explanation

This code uses a **while loop**, which works differently from the for loop in Topic 1:

1. We start by creating a variable `counter` and setting it to 1.
2. The `while` keyword checks a condition: "Is counter less than or equal to 5?" If yes, it runs the code inside the block.
3. After printing the Customer ID, we manually increase `counter` by 1.
4. The loop goes back to check the condition again. Once `counter` reaches 6, the condition becomes false and the loop stops.

**Key takeaway:** A while loop keeps running as long as its condition is true. You control when it stops by updating the variables involved in the condition.

### Explanation Prompt
> How is this different from the code we wrote earlier to print customer IDs? When would I use this approach instead?

```python
# Comparing both approaches side by side

print("=== Using a for loop (Topic 1) ===")
for i in range(1, 6):
    print(f"cid_{i:04d}")

print("\n=== Using a while loop (Topic 2) ===")
counter = 1
while counter <= 5:
    print(f"cid_{counter:04d}")
    counter += 1
```

### Explanation

Both approaches produce the exact same output, but they differ in how they control repetition:

| Aspect | For Loop | While Loop |
|:---|:---|:---|
| Best used when | You know exactly how many times to repeat | You want to repeat until a condition changes |
| Counter management | Handled automatically by `range()` | You must create and update the counter yourself |
| Risk of infinite loop | Very low | Higher, if you forget to update the counter |

**When to use which:** If Cred Pay knows there are exactly 5 customers, a for loop is simpler. But if the rule is "keep processing until a certain condition is met" (like running out of applications), a while loop is the natural fit.

### Modification Prompt
> Instead of stopping at cid_0005, can you make it stop when it reaches the customer whose credit score is below 500? That would be cid_0004. Just use the customer number for now.

```python
# Stop processing when we reach the customer with a low credit score (customer 4)

counter = 1
stop_at = 4  # cid_0004 has a credit score of 462, which is below 500

while counter < stop_at:
    customer_id = f"cid_{counter:04d}"
    print(f"Processing: {customer_id}")
    counter += 1

print(f"\nStopped before cid_{stop_at:04d} (credit score below 500).")
```

### Explanation

The stopping condition changed from `counter <= 5` to `counter < stop_at`, where `stop_at` is set to 4. This means the loop processes customers 1, 2, and 3, then stops before reaching customer 4.

Right now, we are using a hard coded number for the stop point. Later, when we learn to store actual customer data in structured formats, we will be able to look up real credit scores and make the condition dynamic.

# Topic 3: Lists

### Generation Prompt
> I want to store all five credit scores (650, 723, 581, 462, 773) together in one place in Python. Then, I want the program to go through each score and print it along with its position (1st, 2nd, 3rd, etc.).

```python
# Storing all credit scores in a single list

credit_scores = [650, 723, 581, 462, 773]

# Using a for loop to print each score with its position
for index, score in enumerate(credit_scores, start=1):
    print(f"Position {index}: Credit Score = {score}")
```

### Explanation

A **list** is a collection that holds multiple values in a specific order, enclosed in square brackets `[]`.

Here is what the code does:

1. `credit_scores = [650, 723, 581, 462, 773]` stores all five scores in one variable instead of creating five separate ones.
2. `enumerate(credit_scores, start=1)` goes through the list and provides both the position number and the value at each step.
3. The for loop (from Topic 1) handles the repetition automatically.

**Key takeaway:** Lists let you group related data together. This makes it easy to loop through, search, sort, or modify the data without managing individual variables.

### Explanation Prompt
> Can you explain what a list is and why it's better than creating five separate variables for each score?

```python
# Without a list: five separate variables
score_1 = 650
score_2 = 723
score_3 = 581
score_4 = 462
score_5 = 773

print("Without a list, printing all scores requires writing each one manually:")
print(score_1, score_2, score_3, score_4, score_5)

print()

# With a list: one variable holds everything
credit_scores = [650, 723, 581, 462, 773]

print("With a list, you can loop through all scores in two lines:")
for score in credit_scores:
    print(score)
```

### Explanation

When you use separate variables, every operation (printing, searching, sorting) must be written out for each variable individually. If Cred Pay grows from 5 to 500 customers, you would need 500 variables and 500 print statements.

With a list, the same loop works whether you have 5 scores or 5,000. You can also use built in operations like `len(credit_scores)` to get the count, `sum(credit_scores)` to get the total, or `credit_scores.sort()` to arrange them in order.

### Modification Prompt
> A new customer cid_0006 has a credit score of 690. Can you add it to the list and print all scores? This time, can you use the condition based approach from Topic 2 to go through the list instead?

```python
# Adding a new credit score to the list
credit_scores = [650, 723, 581, 462, 773]
credit_scores.append(690)  # Adding cid_0006's score

print(f"Updated list: {credit_scores}")
print(f"Total scores: {len(credit_scores)}")
print()

# Using a while loop to go through the list
index = 0

while index < len(credit_scores):
    print(f"Position {index + 1}: Credit Score = {credit_scores[index]}")
    index += 1
```

### Explanation

Two new concepts appear here:

1. **`.append(690)`** adds the new score to the end of the list. The list grows from 5 items to 6 automatically.
2. **`len(credit_scores)`** returns the current length of the list, which the while loop uses as its stopping condition.

Notice how the while loop from Topic 2 is now combined with the list from this topic. The condition `index < len(credit_scores)` ensures the loop adapts automatically, regardless of how many scores are in the list. This is how each topic builds on the ones before it.

# Topic 4: Tuples

### Generation Prompt
> Each customer at Cred Pay has an ID and a debt status that shouldn't change once recorded. Can you store these pairs, like (cid_0001, in_debt), (cid_0002, no_debt), and so on for all five, in Python in a way that prevents accidental changes? Then print each pair.

```python
# Storing customer ID and debt status as tuples (unchangeable pairs)

applicant_1 = ("cid_0001", "in_debt")
applicant_2 = ("cid_0002", "no_debt")
applicant_3 = ("cid_0003", "no_debt")
applicant_4 = ("cid_0004", "in_debt")
applicant_5 = ("cid_0005", "no_debt")

# Printing each pair
print(applicant_1)
print(applicant_2)
print(applicant_3)
print(applicant_4)
print(applicant_5)
```

### Explanation

A **tuple** is similar to a list, but with one important difference: once created, its values cannot be changed. Tuples use parentheses `()` instead of square brackets `[]`.

This makes tuples ideal for data that should remain fixed after recording. In the Cred Pay context, a customer's ID and their debt status at the time of application are facts that should not be accidentally overwritten.

If you try to change a value inside a tuple (for example, `applicant_1[0] = "new_id"`), Python will raise an error. This is a built in safety feature.

### Explanation Prompt
> What's the difference between storing data the way we did for credit scores versus this new approach? When should I use which?

```python
# Demonstrating the difference between lists (changeable) and tuples (fixed)

# List: you CAN change values
scores_list = [650, 723, 581]
print(f"Original list: {scores_list}")
scores_list[0] = 999  # This works fine
print(f"Modified list: {scores_list}")

print()

# Tuple: you CANNOT change values
applicant_tuple = ("cid_0001", "in_debt")
print(f"Original tuple: {applicant_tuple}")
try:
    applicant_tuple[0] = "cid_9999"  # This will cause an error
except TypeError as e:
    print(f"Error: {e}")
    print("Tuples do not allow changes after creation.")
```

### Explanation

| Feature | List | Tuple |
|:---|:---|:---|
| Syntax | Square brackets `[]` | Parentheses `()` |
| Can be changed after creation? | Yes | No |
| Best for | Data that may need updates (e.g., credit scores) | Data that should stay fixed (e.g., recorded IDs, snapshots) |

Think of a list as a whiteboard where you can erase and rewrite, and a tuple as a printed receipt that cannot be altered once issued.

### Modification Prompt
> Can you put all five of these pairs into a single collection, so I have everything in one place? Then go through it and print the Customer ID and Debt Status on separate lines for each customer.

```python
# Storing all five tuples inside a single list

applicants = [
    ("cid_0001", "in_debt"),
    ("cid_0002", "no_debt"),
    ("cid_0003", "no_debt"),
    ("cid_0004", "in_debt"),
    ("cid_0005", "no_debt")
]

# Looping through the list and unpacking each tuple
for cid, debt_status in applicants:
    print(f"Customer ID   : {cid}")
    print(f"Debt Status   : {debt_status}")
    print()
```

### Explanation

This code introduces a **list of tuples**, which is one of the most common patterns in Python for working with tabular data.

1. The outer structure is a list `[]`, which allows us to loop through all applicants.
2. Each item inside the list is a tuple `()`, which keeps the ID and debt status locked together as a fixed pair.
3. `for cid, debt_status in applicants:` automatically splits each tuple into two named variables on every pass. This is called **unpacking**.

Think of it like the original data table: each row is a tuple, and the full table is the list.

# Topic 5: Dictionaries

### Generation Prompt
> I want to store each Cred Pay customer's complete details (credit score, age, debt status, and number of existing credit cards) in Python. I should be able to look up any customer by their ID and instantly see all their information. Can you set this up for all five customers and print everyone's details?

```python
# Storing complete customer records using a dictionary

applicants = {
    "cid_0001": {"credit_score": 650, "age": 25, "debt_status": "in_debt", "num_cards": 3},
    "cid_0002": {"credit_score": 723, "age": 28, "debt_status": "no_debt", "num_cards": 2},
    "cid_0003": {"credit_score": 581, "age": 38, "debt_status": "no_debt", "num_cards": 4},
    "cid_0004": {"credit_score": 462, "age": 41, "debt_status": "in_debt", "num_cards": 1},
    "cid_0005": {"credit_score": 773, "age": 62, "debt_status": "no_debt", "num_cards": 2}
}

# Printing all customer details
for cid, details in applicants.items():
    print(f"Customer: {cid}")
    print(f"  Credit Score : {details['credit_score']}")
    print(f"  Age          : {details['age']}")
    print(f"  Debt Status  : {details['debt_status']}")
    print(f"  Credit Cards : {details['num_cards']}")
    print()
```

### Explanation

A **dictionary** stores data as key and value pairs, enclosed in curly braces `{}`. In this case:

1. Each **key** is a Customer ID (e.g., `"cid_0001"`), and each **value** is another dictionary containing that customer's details.
2. You can look up any customer instantly by their ID: `applicants["cid_0003"]` returns all of cid_0003's information.
3. `.items()` lets the for loop access both the key (customer ID) and value (details) on each pass.

**Why this is an upgrade from earlier topics:**
Previously, lists stored raw numbers and tuples stored fixed pairs. Dictionaries store complete, labelled records. You no longer need to remember that "credit score is at position 0." You simply ask for it by name: `details['credit_score']`.

### Explanation Prompt
> Can you explain how this way of storing data is different from what we used before for credit scores and customer pairs? Why is this better for complete customer records?

```python
# Quick comparison: accessing data with a list vs a dictionary

# With a list, you access by position (index)
scores = [650, 723, 581, 462, 773]
print(f"Third customer's score (list): {scores[2]}")
print("You need to remember that index 2 = third customer\n")

# With a dictionary, you access by a meaningful label
applicants = {
    "cid_0001": {"credit_score": 650, "age": 25, "debt_status": "in_debt", "num_cards": 3},
    "cid_0002": {"credit_score": 723, "age": 28, "debt_status": "no_debt", "num_cards": 2},
    "cid_0003": {"credit_score": 581, "age": 38, "debt_status": "no_debt", "num_cards": 4},
    "cid_0004": {"credit_score": 462, "age": 41, "debt_status": "in_debt", "num_cards": 1},
    "cid_0005": {"credit_score": 773, "age": 62, "debt_status": "no_debt", "num_cards": 2}
}

print(f"cid_0003's score (dict): {applicants['cid_0003']['credit_score']}")
print(f"cid_0003's age (dict):   {applicants['cid_0003']['age']}")
print("You ask for exactly what you need by name")
```

### Explanation

| Feature | List | Tuple | Dictionary |
|:---|:---|:---|:---|
| Access method | By position number | By position number | By a meaningful key (name/ID) |
| Can be changed? | Yes | No | Yes |
| Best for | Ordered collections of similar items | Fixed, unchangeable records | Labelled records with multiple fields |

Dictionaries are the best fit for customer records because each Customer ID is unique (just like in a real database), and each field has a clear name. You never have to guess which position holds which piece of data.

This is exactly what Cred Pay would do to shortlist eligible applicants: loop through all records, apply a rule, and extract the ones that qualify.

### Validation Prompt
> Before adding a new customer, I want the program to check that: the credit score is between 300 and 900, the age is at least 18, and the debt status is either "in_debt" or "no_debt." If anything is wrong, it should show an error message instead of saving the record. Can you add these safety checks?

```python
# Applicant dictionary
applicants = {
    "cid_0001": {"credit_score": 650, "age": 25, "debt_status": "in_debt", "num_cards": 3},
    "cid_0002": {"credit_score": 723, "age": 28, "debt_status": "no_debt", "num_cards": 2},
    "cid_0003": {"credit_score": 581, "age": 38, "debt_status": "no_debt", "num_cards": 4},
    "cid_0004": {"credit_score": 462, "age": 41, "debt_status": "in_debt", "num_cards": 1},
    "cid_0005": {"credit_score": 773, "age": 62, "debt_status": "no_debt", "num_cards": 2}
}

# Adding a new customer with validation checks
new_cid = "cid_0006"
new_score = 690
new_age = 35
new_debt = "no_debt"
new_cards = 2

# Validation
is_valid = True

if new_score < 300 or new_score > 900:
    print(f"Error: Credit score {new_score} is out of the valid range (300 to 900).")
    is_valid = False

if new_age < 18:
    print(f"Error: Age {new_age} is below the minimum required age of 18.")
    is_valid = False

if new_debt not in ["in_debt", "no_debt"]:
    print(f"Error: Debt status '{new_debt}' is not recognized. Use 'in_debt' or 'no_debt'.")
    is_valid = False

# Only add if all checks passed
if is_valid:
    applicants[new_cid] = {
        "credit_score": new_score,
        "age": new_age,
        "debt_status": new_debt,
        "num_cards": new_cards
    }
    print(f"Customer {new_cid} added successfully!")
else:
    print("Customer was NOT added due to validation errors.")

print(f"\nTotal customers in records: {len(applicants)}")

# Try with bad data to see validation in action
print("\n--- Testing with invalid data ---")
bad_score = 1200
bad_age = 15
bad_debt = "unknown"

if bad_score < 300 or bad_score > 900:
    print(f"Error: Credit score {bad_score} is out of the valid range (300 to 900).")

if bad_age < 18:
    print(f"Error: Age {bad_age} is below the minimum required age of 18.")

if bad_debt not in ["in_debt", "no_debt"]:
    print(f"Error: Debt status '{bad_debt}' is not recognized.")
```

### Explanation

The code now validates every piece of data before saving it. Three checks are performed:

1. **Credit score range:** Must be between 300 and 900. A score of 1200 or a negative number would be rejected.
2. **Minimum age:** The customer must be at least 18 years old.
3. **Valid debt status:** Only "in_debt" or "no_debt" are accepted. Anything else (like "unknown" or a typo) is flagged.

A flag variable `is_valid` tracks whether all checks passed. Only when every condition is satisfied does the record get added to the dictionary. This is a fundamental practice in data handling: always validate before you store. Bad data leads to bad decisions.

# Topic 6: Basic Functions

### Generation Prompt
> The code is getting long and repetitive. Can you help me organize it by creating separate reusable blocks for the following?
> 1. One block that takes a customer's details and creates their record.
> 2. One block that checks if the data is valid (using the safety rules from before) and adds the customer only if everything is correct.
> 3. One block that takes a minimum credit score and gives back only the customers who meet or exceed that score.
>
> Then use these blocks to set up all five customers and show which ones are eligible with a minimum score of 600.

```python
# ---- Reusable Functions ----

def create_applicant(credit_score, age, debt_status, num_cards):
    """Creates and returns a single applicant record as a dictionary."""
    return {
        "credit_score": credit_score,
        "age": age,
        "debt_status": debt_status,
        "num_cards": num_cards
    }


def add_applicant(applicants, cid, credit_score, age, debt_status, num_cards):
    """Validates data and adds the applicant to the dictionary if all checks pass."""
    if credit_score < 300 or credit_score > 900:
        print(f"Error: Credit score {credit_score} is out of the valid range (300 to 900).")
        return False

    if age < 18:
        print(f"Error: Age {age} is below the minimum required age of 18.")
        return False

    if debt_status not in ["in_debt", "no_debt"]:
        print(f"Error: Debt status '{debt_status}' is not recognized.")
        return False

    applicants[cid] = create_applicant(credit_score, age, debt_status, num_cards)
    return True


def get_eligible_applicants(applicants, min_score):
    """Returns a list of Customer IDs whose credit score meets or exceeds the minimum."""
    eligible = []
    for cid, details in applicants.items():
        if details["credit_score"] >= min_score:
            eligible.append(cid)
    return eligible


# ---- Setting Up All Five Customers ----

applicants = {}

add_applicant(applicants, "cid_0001", 650, 25, "in_debt", 3)
add_applicant(applicants, "cid_0002", 723, 28, "no_debt", 2)
add_applicant(applicants, "cid_0003", 581, 38, "no_debt", 4)
add_applicant(applicants, "cid_0004", 462, 41, "in_debt", 1)
add_applicant(applicants, "cid_0005", 773, 62, "no_debt", 2)

print(f"Total customers added: {len(applicants)}\n")

# ---- Finding Eligible Applicants ----

eligible = get_eligible_applicants(applicants, 600)
print(f"Customers eligible with a minimum score of 600:")
for cid in eligible:
    score = applicants[cid]["credit_score"]
    print(f"  {cid} | Score: {score}")
```

### Explanation

The code is now organized into three **functions**, each handling a specific task:

| Function | What It Does |
|:---|:---|
| `create_applicant()` | Takes customer details as inputs and returns a structured record (dictionary) |
| `add_applicant()` | Runs validation checks first, then uses `create_applicant()` to build and store the record |
| `get_eligible_applicants()` | Loops through all records and collects those meeting the minimum score |

**Why this matters:**
Previously, the validation code and record creation code were written directly in the main flow. If you needed to add 100 customers, you would copy and paste the same block 100 times. Now, adding a customer is a single function call: `add_applicant(...)`. If a bug is found in the validation logic, you fix it once inside the function and it is fixed everywhere.

### Explanation Prompt
> Why is organizing code into these reusable blocks helpful? What if Cred Pay grows from 5 customers to 5,000?

### Modification Prompt
> Can you create one more reusable block that goes through all customers and gives back a quick summary: total number of customers, the average credit score, and how many are in debt? Then print the summary.

```python
# New function: get a quick summary of all applicants

def get_applicant_summary(applicants):
    """Returns a summary as a tuple: (total_customers, average_score, count_in_debt)."""
    total = len(applicants)

    if total == 0:
        return (0, 0, 0)

    total_score = 0
    in_debt_count = 0

    for cid, details in applicants.items():
        total_score += details["credit_score"]
        if details["debt_status"] == "in_debt":
            in_debt_count += 1

    average_score = round(total_score / total, 2)

    return (total, average_score, in_debt_count)


# ---- Building the applicant records ----

applicants = {}
add_applicant(applicants, "cid_0001", 650, 25, "in_debt", 3)
add_applicant(applicants, "cid_0002", 723, 28, "no_debt", 2)
add_applicant(applicants, "cid_0003", 581, 38, "no_debt", 4)
add_applicant(applicants, "cid_0004", 462, 41, "in_debt", 1)
add_applicant(applicants, "cid_0005", 773, 62, "no_debt", 2)

# ---- Getting and printing the summary ----

total, avg_score, debt_count = get_applicant_summary(applicants)

print("Cred Pay Applicant Summary")
print("=" * 35)
print(f"Total Customers    : {total}")
print(f"Average Credit Score: {avg_score}")
print(f"Customers in Debt  : {debt_count}")
```

### Explanation

The new function `get_applicant_summary()` ties together concepts from nearly every topic covered so far:

1. It **loops** through the dictionary (Topics 1 and 5).
2. It accumulates totals using variables updated on each pass (Topic 2's counter concept).
3. It **returns a tuple** (Topic 4) containing three values: total customers, average score, and debt count.

The line `total, avg_score, debt_count = get_applicant_summary(applicants)` unpacks the returned tuple into three named variables (just like in Topic 4). The function does all the math internally and hands back a clean, ready to use summary.

### Validation Prompt
> Can you update the safety check block so it also verifies that: the Customer ID follows the format "cid_" followed by exactly four digits, the Customer ID doesn't already exist in our records, and the number of credit cards is not a negative number? It should clearly tell the user what went wrong if any check fails.

```python
# Enhanced add_applicant function with additional validation

def add_applicant(applicants, cid, credit_score, age, debt_status, num_cards):
    """
    Validates all inputs and adds the applicant only if every check passes.
    Returns True if added successfully, False otherwise.
    """

    # Check 1: Customer ID format (must be "cid_" followed by exactly 4 digits)
    if not cid.startswith("cid_") or len(cid) != 8 or not cid[4:].isdigit():
        print(f"Error: '{cid}' is not a valid Customer ID. Expected format: cid_XXXX (four digits).")
        return False

    # Check 2: Customer ID must not already exist
    if cid in applicants:
        print(f"Error: Customer '{cid}' already exists in the records.")
        return False

    # Check 3: Credit score must be between 300 and 900
    if credit_score < 300 or credit_score > 900:
        print(f"Error: Credit score {credit_score} is out of the valid range (300 to 900).")
        return False

    # Check 4: Age must be at least 18
    if age < 18:
        print(f"Error: Age {age} is below the minimum required age of 18.")
        return False

    # Check 5: Debt status must be recognized
    if debt_status not in ["in_debt", "no_debt"]:
        print(f"Error: Debt status '{debt_status}' is not recognized. Use 'in_debt' or 'no_debt'.")
        return False

    # Check 6: Number of credit cards cannot be negative
    if num_cards < 0:
        print(f"Error: Number of credit cards cannot be negative. Received: {num_cards}.")
        return False

    # All checks passed
    applicants[cid] = create_applicant(credit_score, age, debt_status, num_cards)
    print(f"Customer {cid} added successfully.")
    return True


# ---- Testing with various inputs ----

applicants = {}

print("=== Adding valid customers ===\n")
add_applicant(applicants, "cid_0001", 650, 25, "in_debt", 3)
add_applicant(applicants, "cid_0002", 723, 28, "no_debt", 2)

print("\n=== Testing invalid inputs ===\n")
add_applicant(applicants, "abc_9999", 700, 30, "no_debt", 1)     # Bad ID format
add_applicant(applicants, "cid_0001", 700, 30, "no_debt", 1)     # Duplicate ID
add_applicant(applicants, "cid_0003", 1200, 30, "no_debt", 1)    # Score too high
add_applicant(applicants, "cid_0004", 700, 15, "no_debt", 1)     # Too young
add_applicant(applicants, "cid_0005", 700, 30, "unknown", 1)     # Bad debt status
add_applicant(applicants, "cid_0006", 700, 30, "no_debt", -2)    # Negative cards

print(f"\nFinal record count: {len(applicants)}")
```

### Explanation

The enhanced function now performs six validation checks before allowing any data to be stored:

| Check | Rule | Example of Bad Input |
|:---|:---|:---|
| 1. ID format | Must start with "cid_" followed by exactly 4 digits | "abc_9999" |
| 2. Duplicate check | ID must not already exist in the records | Adding "cid_0001" twice |
| 3. Credit score range | Must be between 300 and 900 | 1200 |
| 4. Minimum age | Must be at least 18 | 15 |
| 5. Valid debt status | Must be "in_debt" or "no_debt" | "unknown" |
| 6. Non negative cards | Cannot be a negative number | Negative 2 |

Each check provides a clear, specific error message so the user knows exactly what went wrong. The function returns `True` or `False` so the rest of the program can respond accordingly.

This is how real world software evolves: validation rules grow over time as new edge cases are discovered, and keeping all checks inside a single function ensures they are applied consistently every time.
